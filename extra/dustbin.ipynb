```python
import shutil
import os

# Correct folder path
folder_path = "/home/infodna/Court-MOE/LegalTech"

print(f"Attempting to delete: {folder_path}")

if os.path.exists(folder_path):
    try:
        shutil.rmtree(folder_path, ignore_errors=True)
        print(f"Folder {folder_path} has been deleted.")
    except PermissionError:
        print(f"Permission denied to delete {folder_path}. Try running with sudo or adjust permissions.")
    except Exception as e:
        print(f"An error occurred: {e}")
else:
    print(f"Folder {folder_path} does not exist or path is incorrect.")
```
